```python
top_batsman=df.groupby(['batsman'])['total_runs'].sum().reset_index()
top_batsman
```

```python
top_batsman=top_batsman.sort_values(by=['total_runs'],ascending=False)
top_batsman.head(10)
```

```python
runs=df.groupby(['batsman'])['batsman_runs'].sum().reset_index()
runs.columns=['batsman','batsman_runs']
runs
```

```python
balls=df.groupby(['batsman'])['ball'].count().reset_index()
balls.columns=['batsman','balls']
balls
```

```python
four=df[df['batsman_runs']==4]
four
```

```python
six=df[df['batsman_runs']==6]
six
```

```python
runs_4=four.groupby('batsman')['batsman_runs'].count().reset_index()
runs_4.columns=['batsman','4s']
runs_4
```

```python
runs_6=six.groupby('batsman')['batsman_runs'].count().reset_index()
runs_6.columns=['batsman','6s']
runs_6
```

```python
player=pd.concat([runs,balls.iloc[:,1],runs_4.iloc[:,1],runs_6.iloc[:,1]],axis=1)
player 
```

```python
player.fillna(0,inplace=True)
player
```

```python
player['strike_rate']=(player['batsman_runs']/player['balls'])*100
player
```

```python
grp=df.groupby(['match_id','batsman','batting_team'])['batsman_runs'].sum().reset_index()
grp
```

```python
max_by_each=grp.groupby('batsman')['batsman_runs'].max().reset_index()
max_by_each.columns=['batsman','max_runs']
max_by_each
```

```python
player2=pd.concat([player,max_by_each.iloc[:,1]],axis=1)
player2
```
