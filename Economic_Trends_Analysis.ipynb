```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv('Global Economy Indicators.csv')
data
```

```python
print(data.isnull().sum())
```

```python
cleaned_data=data.dropna()
cleaned_data
```

```python
print(cleaned_data.info())
```

```python
cleaned_data.columns = ['countryId', 'country','year','AMA_exchangeRate','IMF_exchangeRate','population','currency','perCapitaGNI','ISIC_AB','changeInInventories','ISIC_F','exportsGS','finalExpenditure','govtFinalExpenditure','GCF','GFCF','HCE','importsGS','ISIC_D','ISIC_CE','ISIC_JP','totalValue','ISIC_I','ISIC_GH','GNI','GDP']
cleaned_data
```

```python
print("\nSummary Statistics:")
print(cleaned_data.describe().T.to_markdown())
```

```python
px.line(cleaned_data,x='year', y='GDP', color='country',title='GDP Trends Over Time').show()
```

```python
px.scatter(cleaned_data, x='year', y='GNI', color='country', title='GNI Trends Over Time').show()
```

```python
px.line(cleaned_data, x='year', y='AMA_exchangeRate', color='country', title='AMA Exchange Rate Trends Over Time').show()
```

```python
px.line(cleaned_data, x='year', y='perCapitaGNI', color='country', title='AMA Exchange Rate Trends Over Time').show()
```

```python
sectors = ['ISIC_AB', 'ISIC_D', 'exportsGS']
for sector in sectors:
    fig = px.bar(cleaned_data, x='year', y=sector, color='country', title=f'{sector} Contribution Over Time')
    fig.show()
```
